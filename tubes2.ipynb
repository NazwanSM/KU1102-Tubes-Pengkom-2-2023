```python
df["Locality"].value_counts().plot(kind = "bar", title="Locality", color="#6497b1", legend=True)
plt.xlabel("Locality")
plt.ylabel("Frequency")
# menggunakan warna ini agar terlihat kontras dengan warna background putih dan hitam dan agar terlihat menarik
# memilih bar karena masing-masing locality/wilayah nya terlihat
```

```python
df["Network Type"].value_counts().plot(kind="line", title="Network Type", color="#6497b1", legend=True)
plt.xlabel("Network Type")
plt.ylabel("Frequency")
# menggunakan warna ini agar terlihat kontras dengan warna background putih dan hitam dan agar terlihat menarik
# memilih line agar terlihat jelas perbedaan tiap-tiap tipe jaringannya
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Days'] = df['Timestamp'].dt.day
df['Days'].value_counts().sort_index().plot(kind='bar', color='#6497b1', title='Kuantitas terhadap Hari', xlabel='Hari ke-', ylabel='Jumlah Data', legend=True)
# menggunakan warna ini agar terlihat kontras dengan warna background putih dan hitam dan agar terlihat menarik
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Month'] = df['Timestamp'].dt.month
df['Month'].value_counts().sort_index().plot(kind='bar', color='#6497b1', title='Kuantitas terhadap Bulan', xlabel='Bulan ke-', ylabel='Jumlah Data', legend=True)
# menggunakan warna ini agar terlihat kontras dengan warna background putih dan hitam dan agar terlihat menarik
```
